```python
#Use the interquartile range to remove the outliers from non fraud
Q1 = df_nonFraud.quantile(q=.25)
Q3 = df_nonFraud.quantile(q=.75)
IQR = df_nonFraud.apply(stats.iqr)


data_clean_non = df_nonFraud[~((df_nonFraud < (Q1-1.5*IQR)) | (df_nonFraud > (Q3+1.5*IQR))).any(axis=1)]


data_clean_non.shape
```

```python
#scatter plot to show outliers in non fraud dataframe after outliers cleaninig
fig, ax = plt.subplots(figsize=(10,8))
plt.scatter(data_clean_non.Time, data_clean_non.Amount, alpha= 0.2, c= 'g')
plt.ylabel('Amount')
plt.xlabel('Time')
plt.title('data_clean_non')
plt.plot()
```

```python
#Use the interquartile range to remove the outliers from fraud
Q1 = df_fraud.quantile(q=.25)
Q3 = df_fraud.quantile(q=.75)
IQR = df_fraud.apply(stats.iqr)


data_clean_fraud = df_fraud[~((df_fraud < (Q1-1.5*IQR)) | (df_fraud > (Q3+1.5*IQR))).any(axis=1)]


data_clean_fraud.shape
```

```python
#scatter plot to show outliers in fraud dataframe after outliers cleaninig
fig, ax = plt.subplots(figsize=(10,8))
plt.scatter(data_clean_fraud.Time, data_clean_fraud.Amount, alpha= 0.2, c= 'g')
plt.ylabel('Amount')
plt.xlabel('Time')
plt.title('data_clean_fraud')
plt.plot()
```

```python
#balance dataset by using sample from non fraud dataframe
data_clean_non = data_clean_non.sample(data_clean_fraud.shape[0])
data_clean_non.shape
```

```python
data = data_clean_fraud.append(data_clean_non,ignore_index=True)
data
```

```python
#compare after balancing the dataset
pd.value_counts(data['Class']).plot.bar()
plt.title('Fraud class histogram')
plt.xlabel('Class')
plt.ylabel('count')
data['Class'].value_counts()
```

PCA:
------------

```python
#all feacures in X and only class in y
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
```

```python
#use normalization to put the data between 0 and 1
sc = MinMaxScaler(feature_range = (0, 1))
X = sc.fit_transform(X)
```

```python
#split the data for test and train
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
#use pca model to fit
pca = PCA().fit(X)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance');
```

```python
#try 
pca = PCA(n_components = 10)

pct = pca.fit_transform(X)

pca_df = pd.DataFrame(pct,columns=['pc1','pc2','pc3','pc4','pc5','pc6','pc7','pc8','pc9','pc10'])

X_df = pd.concat([pca_df,data[['Class']]],axis=1)
```

```python
X_df
```

MLP:
-------------

```python
mlp = Sequential(layers = None, name = None)
mlp.add(Dense(units = 10, activation = 'relu'))
mlp.add(Dense(units = 1, activation = 'sigmoid'))
mlp.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
history = mlp.fit(X_train, y_train,batch_size=30,epochs=20,verbose=1,validation_data=(X_test, y_test))
```

```python
y_pred = mlp.predict(X_test)
y_pred = (y_pred>0.5) 
#print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), 1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
print(history.history.keys())
```

```python
plt.style.use('ggplot')

# Plot training & validation accuracy values
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```
